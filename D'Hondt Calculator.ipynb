```python
import pandas as pd

race = '2016_pres'
df = pd.read_csv(race + '.csv')

col = list(df.columns)
col = col[2:-1]

final_dict = {'State':[], 'Num Seats': []}

for c in col:
    final_dict.update({c: []})

for index, row in df.iterrows():
    nm = row['state']
    seats = row['num_seats']
    
    left = seats

    votes = {'Votes': {}, 'Value': {}, 'Seats': {}}
    quot = []
    
    for c in col:
        votes['Votes'].update({c: row[c]})
        votes['Value'].update({c: row[c]})
        votes['Seats'].update({c: 0})
        quot.append(row[c])
        
    #print('\n------' + nm + '------')
    while left > 0:
        party = max(votes['Value'], key=lambda key: votes['Value'][key])
        votes['Seats'][party] +=1
        
        #print(str(party) + ', ' + str(votes['Value']))
        left -= 1
        try:
            votes['Value'][party] = votes['Votes'][party]/(votes['Seats'][party]+1)
        except:
            print('Make sure votes are in number format')
        
    final_dict['State'].append(nm)
    final_dict['Num Seats'].append(seats)
    
    for c in col:
        final_dict[c].append(votes['Seats'][c])
        
    #print(nm + ': ' + str(votes['Seats']))
        
final_df = pd.DataFrame(final_dict)
final_df.to_csv(race + '_dhondt.csv', index=False) 
    
final_df
```
